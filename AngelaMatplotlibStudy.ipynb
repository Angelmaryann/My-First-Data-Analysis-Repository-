```python
import pandas as pd

import matplotlib.pyplot as plt
```

**Matplotlib Library:**

Matplotlib is a widely used plotting library in Python that provides tools for creating various types of plots and visualizations.

```python
air_quality = pd.read_csv("/content/air_quality_no2.csv", index_col=0, parse_dates=True)
air_quality.head()
```

This code is using the Pandas library in Python to read a CSV file named "air_quality_no2.csv" and store the data in a DataFrame named `air_quality`. Let's break down the code:

This line performs the following actions:

1. **`pd.read_csv()`:** This function is used to read data from a CSV file into a Pandas DataFrame.

2. **File Path:** `"/content/air_quality_no2.csv"` is the file path where the CSV file is located. You need to replace this path with the actual path to your file.

3. **`index_col=0`:** This parameter specifies that the data in the first column of the CSV file should be used as the index (row labels) of the DataFrame. The index is essentially a way to uniquely identify each row.

4. **`parse_dates=True`:** This parameter tells Pandas to attempt to parse the dates in the CSV file and convert them into a datetime format. This is useful when the data contains date/time information, and you want to work with it as datetime objects in your DataFrame.

After running this code, the `air_quality` DataFrame will contain the data from the CSV file, with the first column used as the index and any date columns parsed into datetime objects. You can then analyze, manipulate, and visualize the data using Pandas functions and methods.

**To plot each of the columns in a separate subplot.**

```python
axs = air_quality.plot.area(figsize=(14, 6), subplots=True)
```

This line of code is using the Pandas plotting functionality to create an area plot from the `air_quality` DataFrame and assigning the resulting plot(s) to the variable `axs`. Let's break down the code:

1. **`air_quality.plot.area`:**
   - This is using the `area` method of the `plot` attribute of the `air_quality` DataFrame. It creates an area plot, which is a type of plot where the area between the x-axis and the plotted line(s) is filled.

2. **`figsize=(14, 6)`:**
   - This parameter sets the size of the figure (the overall image) in inches. In this case, the width is 14 inches, and the height is 6 inches.

3. **`subplots=True`:**
   - This parameter is set to `True`, indicating that separate subplots should be created for each column in the DataFrame. Each column of data will be plotted on its own subplot.

4. **`axs`:**
   - The variable `axs` is assigned the output of the `plot.area` method. This variable will contain one or more AxesSubplot objects, depending on the number of columns in the `air_quality` DataFrame.

The result of this line of code is a set of area plots, one for each column in the DataFrame, arranged as subplots. The `axs` variable can be used to further customize or manipulate the individual subplots if needed. If you run this code in a Jupyter Notebook or a similar interactive environment, you should see the area plots displayed.

**To further customize, extend or save the resulting plot.**

```python
fig, axs = plt.subplots(figsize=(14, 6))  # Create an empty Matplotlib Figure and Axes
air_quality.plot.area(ax=axs)  # Use pandas to put the area plot on the prepared Figure/Axes
```

This code is using the Matplotlib library to create a subplot (a set of axes) and then plotting an area plot of the `air_quality` DataFrame on that subplot. Let's break down the code:

1. **`plt.subplots(figsize=(14, 6))`:**
   - This line creates a subplot (axes) for plotting. The `subplots` function is used to create a figure (`fig`) and one or more subplots (`axs`).
  
   - `fig, axs` is a tuple where `fig` is the entire figure, and `axs` is an array of AxesSubplot objects (in this case, a single subplot).
   - The `figsize=(14, 6)` parameter sets the size of the entire figure to be 14 inches in width and 6 inches in height.

2. **`air_quality.plot.area(ax=axs)`:**
   - This line plots an area plot on the subplot created in the previous step.
   - `ax=axs` specifies that the plot should be drawn on the subplot represented by the `axs` object.

In summary, the code creates a figure with a single subplot, specifies the size of the figure, and then plots an area plot of the `air_quality` DataFrame on that subplot. The resulting visualization should show the area plot within the specified subplot dimensions. If you run this code, you will likely see a graphical representation of the area plot displayed in the specified figure size.

 **To express the concentration of the station in London in mg/m.**

(If we assume temperature of 25 degrees Celsius and pressure of 1013 hPa, the conversion factor is 1.882)

```python
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
air_quality.head()
```

This code adds a new column to the air_quality DataFrame named "london_mg_per_cubic" and assigns values to it based on a calculation involving the existing "station_london" column. The code creates a new column in the DataFrame named "london_mg_per_cubic."
The values in this new column are calculated by multiplying each corresponding value in the existing "station_london" column by the constant factor 1.882.

To create a new column, use the [] brackets with the new column name at the left side of the assignment.

Note: The calculation of the values is done element-wise. This means all values in the given column are multiplied by the value 1.882 at once. You do not need to use a loop to iterate each of the rows!

**To check the ratio of the values in Paris versus Antwerp and save the result in a new column.**

```python
air_quality["ratio_paris_antwerp"] = (air_quality["station_paris"] / air_quality["station_antwerp"])
air_quality.head()
```

**To rename the data columns to the corresponding station identifiers used by OpenAQ.**

```python
air_quality_renamed = air_quality.rename(columns = {"station_antwerp": "BETR801","station_paris": "FR04014","station_london": "London Westminster",})
air_quality_renamed.head()
```

The rename() function can be used for both row labels and column labels. Provide a dictionary with the keys the current names and the values the new names to update the corresponding names.

The mapping should not be restricted to fixed names only, but can be a mapping function as well. For example, converting the column names to lowercase letters can be done using a function as well:

```python
air_quality_renamed = air_quality_renamed.rename(columns=str.lower)

air_quality_renamed.head()
```
