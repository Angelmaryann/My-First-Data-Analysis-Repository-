Introduction of loc/iloc Operators:

To handle more complex subset selections, the loc and iloc operators are introduced.
loc is used for label-based indexing, and iloc is used for integer-based indexing.
Usage of loc/iloc with Selection Brackets:

When using loc or iloc, you place them in front of the selection brackets ([]).
The part before the comma inside the brackets represents the rows you want to select.
The part after the comma inside the brackets represents the columns you want to select.
Specification of Rows and Columns:

The explanation guides on how to specify rows and columns using these operators.
For columns, you can use column names, row labels, or a condition expression.
The part before the comma (rows) and after the comma (columns) can take various forms:
A single label
A list of labels
A slice of labels
A conditional expression
A colon (:) to specify selecting all rows or columns.
In summary, this guidance is for more advanced data selection scenarios where simple bracket notation may not suffice. The use of loc or iloc operators with selection brackets provides a more flexible and powerful way to select specific subsets of data from a DataFrame.

```python
#To read data in rows 10 till 25 and columns 3 to 5.
titanic.iloc[9:25, 2:5]
```

KEYNOTES TO REMEMBER
>When selecting subsets of data, square brackets [] are used.

>Inside these brackets, you can use a single column/row label, a list of column/row labels, a slice of labels, a conditional expression or a colon.

>Select specific rows and/or columns using loc when using the row and column names.

>Select specific rows and/or columns using iloc when using the positions in the table.

>You can assign new values to a selection based on loc/iloc.

**How to calculate summary statistics**

Aggregating statistics

```python
# What is the average age of the Titanic passengers?
titanic["Age"].mean()
```

```python
# What is the median age and ticket fare price of the Titanic passengers?
titanic[["Age", "Fare"]].median()
```

```python
titanic[["Age", "Fare"]].describe()
```

Instead of the predefined statistics, specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:

```python
titanic.agg({"Age": ["min", "max", "median", "skew"],"Fare": ["min", "max", "median", "mean"],})
```

**Aggregating statistics grouped by category**

```python
titanic[["Age","Sex"]].groupby("Sex").mean()
```

In the previous example, we explicitly selected the 2 columns first. If not, the mean method is applied to each column containing numerical columns by passing numeric_only=True:

```python
titanic.groupby("Sex").mean(numeric_only=True)
```

```python
titanic.groupby("Sex")["Age"].mean()
```

Note in the result in code line 26, the Pclass column contains numerical data but actually represents 3 categories (or factors) with respectively the labels ‘1’, ‘2’ and ‘3’. Calculating statistics on these does not make much sense. Therefore, pandas provides a Categorical data type to handle this type of data.

```python
# What is the mean ticket fare price for each of the sex and cabin class combinations?
titanic.groupby(["Sex","Pclass"])["Fare"].mean()
```

Grouping can be done by multiple columns at the same time. Provide the column names as a list to the groupby() method.

**Count number of records by category**

```python
# What is the number of passengers in each of the cabin classes?
titanic["Pclass"].value_counts()
```

The value_counts() method counts the number of records for each category in a column.

The function is a shortcut, as it is actually a groupby operation in combination with counting of the number of records within each group:

```python
titanic.groupby("Pclass")["Pclass"].count()
```

**Note**

Both size and count can be used in combination with groupby. Whereas size includes NaN values and just provides the number of rows (size of the table), count excludes the missing values. In the value_counts method, use the dropna argument to include or exclude the NaN values.

**REMEMBER**
>Aggregation statistics can be calculated on entire columns or rows.

>groupby provides the power of the split-apply-combine pattern.

>value_counts is a convenient shortcut to count the number of entries in each category of a variable.
